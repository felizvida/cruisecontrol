# Reproduce Tables and Figures

This notebook is a compact guide to the local analysis bundle for the zebrafish telencephalon-versus-tectum paper. It assumes the package root is `examples/codex-zebrafish-brain-paper-14rounds/`.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd().resolve().parents[0]
RESULTS = ROOT / "results"
FIGURES = ROOT / "figures"
ROOT
```

```python
summary = json.loads((RESULTS / "summary_metrics.json").read_text())
key_metrics = {
    "exact_id_jaccard_overlap": summary["source_table_jaccard_overlap"],
    "protein_overlap_fraction": summary["source_table_protein_overlap_fraction"],
    "exact_id_centered_jaccard": summary["exact_id_centered_jaccard"],
    "marker_alignment_fraction": summary["marker_alignment_fraction"],
    "motor_family_exclusion_alignment": summary["motor_family_exclusion_alignment"],
    "acetylation_adjusted_odds_ratio": summary["acetylation_adjusted_odds_ratio"],
}
pd.Series(key_metrics, name="value")
```

```python
table1 = pd.read_csv(RESULTS / "presence_overlap_significance.csv")
table2 = pd.read_csv(RESULTS / "sensitivity_scenarios.csv")
table3 = pd.read_csv(RESULTS / "acetylation_covariate_sensitivity.csv")
table4 = pd.read_csv(RESULTS / "gene_symbol_sensitivity.csv")
table5 = pd.read_csv(RESULTS / "motor_family_breakdown.csv")

display(table1)
display(table2)
display(table3)
display(table4)
display(table5.head(12))
```
